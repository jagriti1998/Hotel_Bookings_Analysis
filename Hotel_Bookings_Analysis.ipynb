## Data Cleaning

```python
#Changing reservation_status_date column type to datetime from object
df.reservation_status_date = pd.to_datetime(df.reservation_status_date)
```

```python
#Summary/descriptive stats of the numeric columns
df.describe()
```

```python
#Summary/descriptive stats of the object datatype columns
df.describe(include='object')
```

```python
#Sum if null values in each column
df.isnull().sum()
```

```python
#Drop the columns 'company' and 'agent' and records having null values in the columns 'children' and 'country'
df.drop(['company', 'agent'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#Eewriting the records in the data frame where the adr value is less than 5000
df = df[df.adr < 5000]
```

```python
df.describe()
```

## Data Analysis and Visualization

```python
#To show the percentage (normalize = True; with help of it) of the canceled and not canceled bookings in the data frame.
#We created new data frame cancelled_per. So, we can use it later on as well
cancelled_per = df['is_canceled'].value_counts(normalize = True)
cancelled_per
```

```python
#Created bar plot to present the count of canceled and not canceled bookings
plt.figure(figsize = (5,5))
plt.title('Reservation Status')
plt.bar(['Not Canceled', 'Canceled'], df.is_canceled.value_counts(), edgecolor= 'k', width=.5);
```

```python
#Created count plot to present the count of canceled and not canceled bookings with respect to resort and city hotel type
plt.figure(figsize = (8,5))
sns.countplot(x='hotel', hue = 'is_canceled', data=df)
plt.title('Reservation Status by Hotel Type')
plt.xlabel('Hotel')
plt.ylabel('Numbers of Reservations')
plt.legend(['Not Canceled', 'Canceled']);
```

```python
#To show the percentage (normalize = True; with help of it) of the canceled and not canceled bookings in the data frame
#We created new data frame resort_hotel with part of resort hotel data only. So, we can use it later on as well
resort_hotel = df[df.hotel == 'Resort Hotel']
resort_hotel.is_canceled.value_counts(normalize=True)
```

```python
#To show the percentage (normalize = True; with help of it) of the canceled and not canceled bookings in the data frame
#We created new data frame city_hotel with part of city hotel data only. So, we can use it later on as well
city_hotel = df[df.hotel == 'City Hotel']
city_hotel.is_canceled.value_counts(normalize=True)
```

```python
#Grouping the dataframe by the reservation_status_date column, and calculating the mean value of the adr column for each group.
#The resulting dataframe is assigned to the variable resort_hotel.
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

#Grouping the dataframe by the reservation_status_date column, and calculating the mean value of the adr column for each group.
#The resulting dataframe is assigned to the variable city_hotel.
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize= (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)

#Plotting the adr column from the resort_hotel dataframe against its index column (which is the reservation_status_date column)
plt.plot(resort_hotel.index, resort_hotel.adr, label ='Resort Hotel')

#Plotting the adr column from the city_hotel dataframe against its index column
plt.plot(city_hotel.index, city_hotel.adr, label='City Hotel')
plt.legend(fontsize=20);
```

```python
df['month'] = df.reservation_status_date.dt.month

plt.figure(figsize = (20,8))

#Plotting number of occurrences of each category in the month column.
#Bars are colored based on the value of the is_canceled column
sns.countplot(x='month', hue='is_canceled', data=df)
plt.title('Reservation Status per Month', size = 20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not canceled', 'Canceled']);
```

```python
plt.figure(figsize=(15,8))

#Shows the total adr values for each month of the year when bookings were canceled
sns.barplot('month', 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.title('ADR per Month');
```

```python
#To show the percentage of the canceled bookings country wise in the data frame.
canceled_df = df[df.is_canceled ==1]
top_10_country = canceled_df.country.value_counts()[0:10]

#To show the percentage of the canceled bookings of top 10 country wise in the data frame with the help of pie chart.
plt.figure(figsize = (5,5))
plt.pie(top_10_country, autopct = '%.2f', labels=top_10_country.index)
plt.title('Top 10 Country with Cancelled Reservation');
```

```python
#To show the value count of the market segment from where bookings were received in the data frame.
df.market_segment.value_counts()
```

```python
#To show the value count of the market segment from where the most cancelations were received in the data frame.
df.market_segment.value_counts(normalize=True)
```

```python
#To show the percentage of the market segment from where the most cancelations were received in the data frame.
canceled_df.market_segment.value_counts(normalize=True)
```

```python
cancelled_df_adr = canceled_df.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_df = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,8))
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.title('Average Daily Rate', fontsize=30);
```

```python
#Taking data for the time period of Jan, 2016 to Sep, 2017 because of continous us available in this time frame
##The resulting dataframe is assigned to the variable cancelled_df_adr and not_cancelled_df_adr respectively.
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))

plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.title('Average Daily Rate', fontsize = 30)
plt.legend(fontsize=15);
```
